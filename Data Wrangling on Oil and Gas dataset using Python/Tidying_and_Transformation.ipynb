# **Tidying and Transformation**

## Tidying+Transforming=Data Wrangling

## Convert dataset from a wider format to a longer format

```
import pandas as pd
#read excel file
untidy_data=pd.read_excel('Production_untidy_data.xlsx')
```

```
untidy_data.head()
```

```
untidy_data.shape
```

```
tidy_data = pd.melt(untidy_data, id_vars=['State'], 
                  var_name='Year', value_name='rate_class_1_oil')
tidy_data.head()
```

```
tidy_data.shape
```

So, the untidy dataset has been transformed from (33,7) to (198,3)

## Now, I want to convert the dataset from longer format to wider format

```
wider_data = tidy_data.pivot_table('rate_class_1_oil', ['State'], 'Year')
wider_data.head()
#here, the transformed dataset has been converted back to it's original format
```

# Organizing the Dataset

```
untidy_1_2=pd.read_excel('Production_untidy_data_reshape_rate_class_1,2_oil.xlsx')
print(untidy_1_2.shape)
print(untidy_1_2.isnull().sum())
```

```
untidy_1_2=untidy_1_2.fillna(0)#fiil the missing values as 0
print(untidy_1_2.isnull().sum())
print(untidy_1_2.shape)
```

```
untidy_1_2.dtypes
```

```
df = pd.melt(untidy_1_2, id_vars=['State', 'Class'], 
                  var_name='Year', value_name='rate_class_oil_value')
print(df.dtypes)
print(df.head())
```

```
df['Year']=df['Year'].astype('int64')#convert Year from object to int64
df['rate_class_oil_value']=df['rate_class_oil_value'].astype('int64')#convert 'rate_class_oil_value' from float64 to int64
print(df.dtypes)
```

```
df.isnull().sum()#again check that there are no missing values in the dataset columns
```

```
df.head()
```

## Let's answer some questions related to the dataset, 'df'

```
## Total no. of wells per state wrt each year and each class
X = df.groupby(['State', 'Year','Class'])['rate_class_oil_value'].sum().to_frame()
X = X.pivot_table('rate_class_oil_value', ['State'], ['Year','Class'])
print(X.shape)
X
```

```
##Total no. of wells per state wrt each year i.e. combine wells in class 1 and 2 for each state
X=df.groupby(['State', 'Year'])['rate_class_oil_value'].sum().to_frame()
X = X.pivot_table('rate_class_oil_value', ['State'], 'Year')
print(X.shape)
X
```

This shows that for Alabama in 2004 there were total 37 wells.

25 'Class 1' wells and 12 'Class 2' wells

```
## Total number of wells per state wrt each class
X=df.groupby(['State', 'Class'])['rate_class_oil_value'].sum().to_frame()
X = X.pivot_table('rate_class_oil_value', ['State'], 'Class')
print(X.shape)
X
```

This shows that there are 144 'Class 1' wells in Alabama and 111 'Class 2' wells
