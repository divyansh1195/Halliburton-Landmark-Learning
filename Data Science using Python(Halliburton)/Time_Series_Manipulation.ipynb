```
#Time Series: Sequence of data points daily, weekly, hourly and quaterly(iscree time data points)
#It predicts future sequence based on previous observed occurances. It understands past trends so that we can forcast and plan for the future.
#Hence, in Time seies Analysis, the only variable is time.
#Time Series accounts for Autocorrelation, which is also affected by earlier observed values.   
```

```
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
from datetime import datetime
```

```
#select a date range
date_rng=pd.date_range(start='1/1/2020', end='31/12/2020', freq='H') #this date range has time stamps with hourly frequency.
date_rng
```

```
len(date_rng)
```

```
#check the first element
type(date_rng[0])
```

```
date_rng[0]
```

**CREATE AN EXAMPLE DATAFRAME WITH TIME STAMP DATA**

```
df=pd.DataFrame(date_rng, columns=['date'])
df['data']=np.random.randint(0,100, size=(len(date_rng)))
df.head(10)
```

**CONVERT THE DATAFRAME INDEX TO DATE TIME INDEX**

```
df['datetime']=pd.to_datetime(df['date'])
df=df.set_index('datetime')
df.drop(['date'], axis=1, inplace=True)
df.head()
```

```
#for time series manipulation dataframe is indexed to date time index
```

**SOMETIMES THE TIME STAMPS IN DATA ARE NOT NUMERICAL**

```
#convert date_rng to a list of strings
string_date_rng=[str(x) for x in date_rng]
string_date_rng
```

```
type(string_date_rng[0])
```

```
#convert the strings to the time stamp
timestamp_date_rng=pd.to_datetime(string_date_rng)
timestamp_date_rng
```

**FILTERING**

```
#we want to observe all the timestamps on the second day of a month
df[df.index.day==2]
```

```
#if we wnat to observe at a particular date
df['2020-2-2'] 
```

```
#we want to observe all the timestamps in 2 particular periods
df['2020-2-2':'2020-6-11']
```

**RESAMPLING**

```
#resample dataset according to different frequencies
df.resample('D').mean()#D-Day, mean of each day 
```

```
df.resample('M').mean()#M-Month, mean of each month
```

**WINDOW STATISTICS-ROLLING MEAN AND ROLLING SUM** 

```
#let's create a new column in our original df that computes the rolling sum over a period window
#take 3 period window
df['rolling_sum']=df.rolling(3).sum()
df.head(10)
```

**HANDLING MISSING DATA**

```
#'backfill' for preceding NaN values
#'forwardfill' for succeding NaN values
df['rolling_sum_backfilled']=df['rolling_sum'].fillna(method='backfill')
df.head(10)
```
