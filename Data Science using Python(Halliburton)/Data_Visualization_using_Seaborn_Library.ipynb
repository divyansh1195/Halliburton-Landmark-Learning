**Data Visualization using Seaborn Library** 

```
#seaborn library is used for drawing statistical graphs
```

```
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
```

```
#tips dataset is already present in sns library
tips=sns.load_dataset("tips")
tips.head()
```

```
#Distribution plot
sns.distplot(tips['total_bill'])
```

```
#decrease the size of the bins
sns.distplot(tips['total_bill'], bins=40)
```

```
#Categorical Plots
#BAR PLOTS
```

```
sns.barplot(x='sex', y='total_bill', data=tips)
```

```
#Countplot to count the total no of male and female in the dataset
sns.countplot(x='sex', data=tips)
```

```
#Box Plot: Min, Max, Median, Ist Quartile, 2nd Quartile
sns.barplot(x='day', y='total_bill', data=tips)
```

**MATRIX PLOT** 

```
#type of dataset
type(tips)
```

```
#correlation matrix
tc=tips.corr()
```

```
tc
```

```
#visualize correlation matrix as a heat map
sns.heatmap(tc, annot= True)
```

**GRID PLOT**


```
#Histogram of each column abd sactter plot of each column wrt other column
sns.pairplot(tips)
```

```
#segration of data between male and female
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```
#Regression Plot using lmplot for plotting 2 regression lines
#what is hue?
#In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding.
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```
