**OBJECT ORIENTED METHOD** 

```
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
#Create a list
X=np.linspace(0,10,100)
y=np.sin(X)
z=np.cos(X)
```

```
list(X)
```

```
list(y)
```

```
list(z)
```

```
#creating graphs using figures
```

```
fig=plt.figure()
ax=fig.add_axes([0.1, 0.1, 1, 1])
ax.plot(X,y, 'r--', lw=2.5, label='sine', ls='--')
ax.plot(X,z, 'b--', alpha=0.39, marker='o', markersize=8, label='cosine')
ax.set_xlim(0, 11)
ax.set_ylim(-1,1)
ax.legend()
ax.set_xlabel('line')
ax.set_ylabel('Amplitude')
ax.set_title('Sin-Cos Graph')
```

```
#creating subplots
```

```
#define both figure and axes 
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(10,5))
#ax(0)#first graph
#ax(1)#second graph
ax[0].plot(X,y, c='brown')
ax[1].plot(X,z, c='green')
```

**Multiple plots in Single Canvas**

```
fig=plt.figure(figsize=(15,9))
ax_1=fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax_2=fig.add_axes([0.35, 0.5, 0.4, 0.1])
ax_1.plot(X,y, 'r--', lw=2.5, label='sine', ls='--')
ax_2.plot(X,z, 'b--', alpha=0.39, marker='o', markersize=8, label='cosine')
ax_1.set_xlabel('Distance')
ax_1.set_ylabel('Amplitude')
ax_1.set_title('Sine Graph')
ax_2.set_xlabel('Distance')
ax_2.set_ylabel('Amplitude')
ax_2.set_title('Cosine Graph')
```

```
fig=plt.figure(figsize=(15,9))
ax_1=fig.add_axes([0.1, 0.1, 0.9, 0.9])
ax_2=fig.add_axes([0.15, 0.15, 0.4, 0.1])
ax_1.plot(X,y, 'r--', lw=2.5, label='sine', ls='--')
ax_2.plot(X,z, 'b--', alpha=0.39, marker='o', markersize=8, label='cosine')
ax_1.set_xlabel('Distance')
ax_1.set_ylabel('Amplitude')
ax_1.set_title('Sine Graph')
ax_2.set_xlabel('Distance')
ax_2.set_ylabel('Amplitude')
ax_2.set_title('Cosine Graph')
```
